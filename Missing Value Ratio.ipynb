## Missing Value Ratio

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("NaN_Housing_Prices.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
### Missing Value Ratio formula ###

# Missing value ratio = number of missing values/Total no.of observations 
```

```python
for i in data.columns:
    tmp = data[i].isnull().sum()/len(data)   ## formula
    if tmp > 0.70:
        print("removing column" + i + "with missing value ratio",tmp)
        data = data.drop(columns = [i])
```
